# Q7 Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

# Q4-  Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd

driver= webdriver.Chrome(r"C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\chromedriver.exe")

url="http://www.flipkart.com"

driver.get(url)

time.sleep(3)

# closing the window frame by sending the escape key 
from selenium.webdriver.common.keys import Keys
driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
time.sleep(3)
search = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
# sending the value to the input "pixel" and clicking on the button
search.send_keys("pixel")
search.find_element_by_xpath('//button').click()

time.sleep(2)
# getting the phone_tag from the class 
phone_tag=driver.find_elements_by_xpath("//div[@class='col col-7-12']")

#creating the list of the all the attributes to be saved
brand=[]
Phone_name=[]
colour=[]
Memory=[]
phone_memory=[]
phone_display=[]
display_size=[]
display_resolution=[]
phone_camera=[]
phone_battery=[]
phone_processor=[]
RAM=[]
primary_camera=[]
secondary_camera=[]
price=[]
product_url=[]
product_url_tag=[]

for phone in phone_tag:
    
    phone_memory=[]
    # Getting the complete text of the name of the phone tag
    phone_name_color=phone.find_element_by_class_name("_4rR01T").text

    # Getting the first value to get the brand name by splitting at the first instance of the space
    brand.append(phone_name_color.split(" ")[0])

    #Splitting at the first instance of "(" to get the phone name 
    Phone_name.append(phone_name_color.split("(")[0])

    #Splitting at the '(' and the ')' and again splitting at "," and getting the colour between two isntances of ( and )
    colour.append(phone_name_color.split('(')[-1].split(")")[0].split(",")[0])
    
    # Splitting the phone memory by getting the text and splitting at delimiter "|" and saving the first value into RAM and second Instance into Memory list
    phone_memory=phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[1]").text
    RAM_HDD=phone_memory.split('|')
    RAM.append(RAM_HDD[0])
    Memory.append(RAM_HDD[1])

    phone_display=phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[2]").text
    # Splitting at delimiter at ")" and adding that value to the display_size list and similarly the second value into display resolution
    display_size.append(phone_display.split(")")[0])
    display_resolution.append(phone_display.split(")")[1])


    phone_camera=phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[3]").text
    try:
        # Splitting the camera text at the delimiter "|" and saving the first part as primary camera and the secondary camera  after the "|"
        primary_camera.append(phone_camera.split("|")[0])
        secondary_camera.append(phone_camera.split("|")[1])
    except IndexError as dex:
        # When no secondary camera returns then append the string "No secondary camera"
        secondary_camera.append("No secondary camera")

    phone_battery.append(phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[4]").text)
    
    try:
        # If it encounter the string "Year" or "year" in the place of battery then just append the values "No Processor Specs"
        if "Year" in phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[5]").text:
            phone_processor.append("No Processor Specs")
        elif "year" in phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[5]").text:
            phone_processor.append("No Processor Specs")
            
        else:
            # else just append the value of the text as the phone processor 
            phone_processor.append(phone.find_element_by_class_name("_1xgFaf").find_element_by_xpath("./li[5]").text)
    except NoSuchElementException as e:
        # if it encounter nosuch element exception then mention no processor specs
        phone_processor.append("No processor spec")

# Appening the value of the price tag to the price list
pricetag=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for cost in pricetag:
    price.append(cost.text)
price.append

# Appending the url by getting the attribute 'href' and appending it to product url
product_url_tag=driver.find_elements_by_class_name("_1fQZEK")
for url in product_url_tag:
    product_url.append(url.get_attribute("href"))

# Just replacing display size value "inch" with "inch)"
for i in range(0,len(display_size)):
    display_size[i]=display_size[i].replace("inch","inch)")
```

# Q3 - Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

driver=webdriver.Chrome(r"C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\chromedriver.exe")
url="http://images.google.com"

driver.get(url)
time.sleep(2)
# Creating a function image with the argument keyword
def image(keyword):
    
    # This part of the code is to scroll till the end of the images window
    last_height = driver.execute_script('return document.body.scrollHeight')
    while True:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        time.sleep(2)
        new_height = driver.execute_script('return document.body.scrollHeight')
        try:
            driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div/div[5]/input').click()
            time.sleep(2)
        except:
            pass
        if new_height == last_height:
            break
        last_height = new_height

    print(keyword)
    # taken the first 100 images of the searched images 
    for i in range(1, 100):
        try:
            # Taking a screenshot of the images searched and assigning the name to them as images(keyword(i).png)
            driver.find_element_by_xpath('//*[@id="islrg"]/div[1]/div['+str(i)+']/a[1]/div[1]/img').screenshot(r'C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\webscraping assignment 3\images\images ('+keyword+str(i)+').png')
        except:
            pass
# USing the for loop to go through 3 keywords 
for i in ["cars","fruits","machine learning"]:
    search_tag=driver.find_element_by_xpath("//input[@title='Search']")
    search_tag.clear()
    search_tag.send_keys(i)
    try:
        # Enter the string is entered and searched using the button
        driver.find_element_by_xpath("//button[@id='BIqFsb']").click()
    except NoSuchElementException as e:
        # at the first instance the string is searched and type button of the value submit is clicked 
        driver.find_element_by_xpath("//button[@type='submit']").click()
    time.sleep(2)
    # Calling the function image by passing the values in the for loop one by one 
    image(i)
```
